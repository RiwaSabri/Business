```python
def landing_page(landing_page):
    if len(landing_page.split('\n')[0].split(','))==76 or len(landing_page.split('\n')[0].split(','))==86:
        return landing_page.split('\n')[0].split(',')[10].split(':')[2][2:-1]
    else: 
        return landing_page.split('\n')[0].split(',')[9].split(':')[2][2:-1]
```

```python
df['Landing_Page']=df.hits.apply(lambda x:landing_page(x))
```
